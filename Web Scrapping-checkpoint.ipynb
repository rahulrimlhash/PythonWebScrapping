#### For getting information from all tables 

```python
for one_table in all_tables:
    all_rows = one_table.find_all('tr')[1:]
    
    for tr in all_rows:
        
        all_tds = tr.find_all('td')
    
        country = all_tds[0].text
        capital = all_tds[1].text
        languages = all_tds[-1].text
        
        ele = {'Country' : country, 'Capital' : capital, 'Languages' : languages}
        df = df.append(ele, ignore_index=True)

        print(country,"  ",capital,"  ",languages)
```

```python
#import pandas as pd

#all_tables = soup.find_all('table')

df = pd.DataFrame(columns=['Country', 'Capital', 'Languages'])

for one_table in all_tables:
    all_rows = one_table.find_all('tr')[1:]

    for tr in all_rows:
        all_tds = tr.find_all('td')

        country = all_tds[0].text
        capital = all_tds[1].text
        languages = all_tds[-1].text

        ele = {'Country' : country, 'Capital' : capital, 'Languages' : languages}
        df = pd.concat([df, pd.DataFrame(ele,index=[0])])

        print(country, "  ", capital, "  ", languages)
```
